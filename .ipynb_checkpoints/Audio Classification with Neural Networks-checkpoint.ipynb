## Data Import

We're going to use the `librosa` package to work with the audio data. [LibROSA](https://librosa.github.io/librosa/) describes itself as a package for music and audio analysis, providing many useful functions for working with audio data and extracting features for use in downstream machine learning tasks. 

Firstly, we'll write a (vectorised) function to load sound files:

```python
def load_sound_files(data_dir, file_names): 
    raw_sounds = []
    for fn in file_names:
        fp = os.path.join(data_dir, fn) 
        X,sr = librosa.load(fp)
        raw_sounds.append(X)
    return raw_sounds
```

```python
# Filename structure is [fold]/[fsID]-[classID]-[occurrenceID]-[sliceID].wav
# [fold]         = Dataset is split into 10 folds (arranged as folders) 
# [fsID]         = the Freesound ID of the recording from which this excerpt (slice) is taken
# [classID]      = a numeric identifier of the sound class (see description of classID below for further details)
# [occurrenceID] = a numeric identifier to distinguish different occurrences of the sound within the original recording
# [sliceID]      = a numeric identifier to distinguish different slices taken from the same occurrence

# Manually pick 10 files to inspect, 1 from each class 
sound_file_paths = ["fold1/57320-0-0-7.wav", "fold1/24074-1-0-3.wav", "fold1/15564-2-0-1.wav", "fold1/31323-3-0-1.wav",
                    "fold1/46669-4-0-35.wav", "fold1/89948-5-0-0.wav", "fold1/40722-8-0-4.wav", "fold1/103074-7-3-2.wav",
                    "fold1/106905-8-0-0.wav", "fold1/108041-9-0-4.wav"]

# We also need to create a lookup for classID
classID = ["air conditioner","car horn","children playing","dog bark","drilling",
           "engine idling", "gun shot","jackhammer","siren","street music"]

# Load the 10 files, and print them out to the notebook to inspect the structure
tmp_sound_files = load_sound_files(Config.data_dir, sound_file_paths)
tmp_sound_files
```

## Feature Engineering

Audio data doesn't immediately lend itself to machine learning (considered naively, 4 seconds of audio at 22050Hz is nearly 100k features), so we need to create some lower-dimensional features to use for learning. There are a few methods we can use here:

* melspectrogram: Compute a Mel-scaled power spectrogram
* mfcc: Mel-frequency cepstral coefficients
* chroma-stft: Compute a chromagram from a waveform or power spectrogram
* spectral_contrast: Compute spectral contrast, using method defined [here](http://ieeexplore.ieee.org/document/1035731/)
* tonnetz: Computes the tonal centroid features (tonnetz), following the method [here](https://dl.acm.org/citation.cfm?id=1178727)

Explanation of what these features mean is beyond the scope of this tutorial, but they're all broadly based around transforming the audio data from the **time domain** to the **frequency domain**.

```python
def extract_feature(data_dir = "/", file_name = ""):
    fp = os.path.join(data_dir, file_name)
    X, sample_rate = librosa.core.load(fp)
    stft = np.abs(librosa.core.stft(X))
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X), sr=sample_rate).T,axis=0)
    return mfccs,chroma,mel,contrast,tonnetz
```

Let's take a look at these features using some air conditioner noise from the samples above. We won't look directly at the features (they're just numbers) but we will look at how they're working to reduce the dimensionality of our data.

```python
print("Length of original audio vector: ", len(tmp_sound_files[0]))
mfccs, chroma, mel, contrast, tonnetz = extract_feature(Config.data_dir, "fold1/57320-0-0-7.wav")
print("Length of MFCC features: ", len(mfccs))
print("Length of chroma-stft features: ", len(chroma))
print("Length of mel-scaled spectrogram features: ", len(mel))
print("Length of spectral contrast features: ", len(contrast))
print("Length of tonnetz features: ", len(tonnetz))
print("Length of combined derived features: ", len(mfccs) + len(chroma) + len(mel) + len(contrast) + len(tonnetz))
```

Using these derived features, we've managed to reduce the feature space from 88200 (which, by the way, may vary across the dataset if the audio files are different lengths) to 193. This is now an appropriate feature space for working with a full-connected neural network.

We now need to calculate these features for all 8732 audio examples in the dataset. This will take a little while to run; it could almost certainly be parallelised, but that's up to you to figure out. I'll give you a progress bar and you can go get a coffee or something.

```python
import progressbar

def parse_audio_files(parent_dir,sub_dirs,file_ext='*.wav'):
    
    # Empty arrays for features (193 columns) and labels (1 column)
    features, labels = np.empty((0,193)), np.empty(0)
    
    for label, sub_dir in enumerate(sub_dirs):
        widgets = [progressbar.FormatLabel(sub_dir), ' ', 
                   progressbar.Percentage(), ' ', 
                   progressbar.Bar('/'), ' ', 
                   progressbar.RotatingMarker()]
        bar = progressbar.ProgressBar(widgets = widgets)
        for fn in bar(glob.glob(os.path.join(parent_dir, sub_dir, file_ext))):
            try: # Some audio files are corrupted - just report errors and move on
                mfccs, chroma, mel, contrast,tonnetz = extract_feature(file_name = fn)
                ext_features = np.hstack([mfccs,chroma,mel,contrast,tonnetz])
                features = np.vstack([features,ext_features])
                labels = np.append(labels, fn.split('/')[-1].split('-')[1])
            except Exception as e:
                print(e)
                print(fn)
    return np.array(features), np.array(labels, dtype = np.int)

def one_hot_encode(labels):
    n_labels = len(labels)
    n_unique_labels = len(np.unique(labels))
    one_hot_encode = np.zeros((n_labels,n_unique_labels))
    one_hot_encode[np.arange(n_labels), labels] = 1
    return one_hot_encode
```

```python
import warnings

train_folds = ['fold1','fold2','fold3','fold4','fold5','fold6','fold7','fold8','fold9']
test_folds = ['fold10']

with warnings.catch_warnings(): # We're going to get some warnings when the spectral contrast filter fails to estimate pitch
    warnings.simplefilter("ignore")
    train_feat, train_labels = parse_audio_files(Config.data_dir, train_folds)
    test_feat, test_labels   = parse_audio_files(Config.data_dir, test_folds)
```

Finally, to prepare the data for training, we need to reshape the labels into a one-hot matrix as they will be training a softmax final layer. We'll also do our own train/development split for model tuning, and then test on fold 10.

```python
# Save these datasets to file to save ever having to run that code again
import pickle

pickle.dump(train_feat,   open(os.path.join(Config.data_dir, "train_feat.pkl"),   "wb"))
pickle.dump(train_labels, open(os.path.join(Config.data_dir, "train_labels.pkl"), "wb"))
pickle.dump(test_feat,    open(os.path.join(Config.data_dir, "test_feat.pkl"),    "wb"))
pickle.dump(test_labels,  open(os.path.join(Config.data_dir, "test_labels.pkl"),  "wb"))
```

```python
# Import the datasets from file if you're running this again
import pickle

train_feat   = pickle.load(open(os.path.join(Config.data_dir, "train_feat.pkl"),   "rb"))
train_labels = pickle.load(open(os.path.join(Config.data_dir, "train_labels.pkl"), "rb"))
test_feat    = pickle.load(open(os.path.join(Config.data_dir, "test_feat.pkl"),    "rb"))
test_labels  = pickle.load(open(os.path.join(Config.data_dir, "test_labels.pkl"),  "rb"))
```

```python
train_labels_1h = one_hot_encode(train_labels)
test_labels_1h  = one_hot_encode(test_labels)

train_test_split = np.random.rand(len(train_feat)) < 0.80
train_x = train_feat[train_test_split]
train_y = train_labels_1h[train_test_split]
devel_x = train_feat[~train_test_split]
devel_y = train_labels_1h[~train_test_split]
test_x = test_feat
test_y = test_labels_1h
```

## Model Training

We will be using Tensorflow to train a fully connected neural network with three hidden layers. The model parameters below were selected based on trial and error - you can try to improve the model by tuning these values if you wish.

```python
import tensorflow as tf
from sklearn.metrics import precision_recall_fscore_support
```

```python
class Params():
    training_epochs      = 50000
    n_dim                = train_feat.shape[1]
    n_classes            = 10
    n_hidden_units_one   = 500 
    n_hidden_units_two   = 400
    n_hidden_units_three = 300
    sd                   = 1 / np.sqrt(n_dim)
    learning_rate        = 0.01
    dropout_keep         = 0.5
```

Here we will define the model using Tensorflow. The model has an input layer, 3 hidden layers (relu, relu, sigmoid), and a softmax output layer (because we're training a classifier).

```python
X = tf.placeholder(tf.float32,[None,Params.n_dim])
Y = tf.placeholder(tf.float32,[None,Params.n_classes])
keep_prob = tf.placeholder(tf.float32)

W_1 = tf.Variable(tf.random_normal([Params.n_dim,Params.n_hidden_units_one], mean = 0, stddev=Params.sd))
b_1 = tf.Variable(tf.random_normal([Params.n_hidden_units_one], mean = 0, stddev=Params.sd))
h_1 = tf.nn.relu(tf.nn.dropout(tf.matmul(X,W_1) + b_1, keep_prob))


W_2 = tf.Variable(tf.random_normal([Params.n_hidden_units_one,Params.n_hidden_units_two], mean = 0, stddev=Params.sd))
b_2 = tf.Variable(tf.random_normal([Params.n_hidden_units_two], mean = 0, stddev=Params.sd))
h_2 = tf.nn.relu(tf.nn.dropout(tf.matmul(h_1,W_2) + b_2, keep_prob))

W_3 = tf.Variable(tf.random_normal([Params.n_hidden_units_two,Params.n_hidden_units_three], mean = 0, stddev=Params.sd))
b_3 = tf.Variable(tf.random_normal([Params.n_hidden_units_three], mean = 0, stddev=Params.sd))
h_3 = tf.nn.sigmoid(tf.nn.dropout(tf.matmul(h_2,W_3) + b_3, keep_prob))


W = tf.Variable(tf.random_normal([Params.n_hidden_units_three,Params.n_classes], mean = 0, stddev=Params.sd))
b = tf.Variable(tf.random_normal([Params.n_classes], mean = 0, stddev=Params.sd))
y_ = tf.nn.softmax(tf.matmul(h_3,W) + b)

init = tf.global_variables_initializer()
```

We're going to use cross-entropy loss with gradient descent. You could probably get the model to converge more quickly if you use `tf.train.AdamOptimizer` but it's throwing errors that I don't have time to debug.

```python
cost_function = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(y_), reduction_indices=[1])) 
optimizer = tf.train.GradientDescentOptimizer(Params.learning_rate).minimize(cost_function)

correct_prediction = tf.equal(tf.argmax(y_,1), tf.argmax(Y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

Now we'll train the model, and produce scores for the development and test datasets.

```python
cost_history = np.empty(shape=[1],dtype=float)
y_true, y_pred = None, None

# Add ops to save and restore all the variables.
saver = tf.train.Saver()

with tf.Session() as sess:
    sess.run(init)
    for epoch in range(Params.training_epochs):            
        _,cost = sess.run([optimizer,cost_function],feed_dict={X:train_x,Y:train_y,keep_prob:Params.dropout_keep})
        cost_history = np.append(cost_history,cost)
    
    y_devel_pred = sess.run(tf.argmax(y_,1),feed_dict={X:devel_x,keep_prob:1.0})
    y_devel_true = sess.run(tf.argmax(devel_y,1))
    y_test_pred = sess.run(tf.argmax(y_,1),feed_dict={X:test_x,keep_prob:1.0})
    y_test_true = sess.run(tf.argmax(test_y,1))
    # Save the variables to disk.
    save_path = saver.save(sess, Config.model_dir + "/model.ckpt")
    print("Model saved in path: %s" % save_path)
```

And now we can plot the cost function to see how the model converged. We'll also take a look at the development and test set performance.

```python
fig = plt.figure(figsize=(10,8))
plt.plot(cost_history)
plt.ylabel("Cost")
plt.xlabel("Iterations")
plt.axis([0,Params.training_epochs,0,np.max(cost_history)])
plt.show()

p,r,f,s = precision_recall_fscore_support(y_devel_true, y_devel_pred, average='micro')
print("Development F-Score:", round(f,3))

p,r,f,s = precision_recall_fscore_support(y_test_true, y_test_pred, average='micro')
print("Test F-Score:", round(f,3))
```

As expected, the development set has unreasonably high accuracy due to the way the data was arranged in the folders. Because many of the clips were longer than 4s, they were broken up into multiple examples. Examples from the same clip are kept in the same "fold", which means that the random test/development split introduced significant leakage. Only the test set (fold10) is a fair estimate of the classifier performance.

But for now, let's see if we can use this model to make some predictions!

```python
test_example = 9 # Any number between 0 and 836 - used to draw an example for scoring

with tf.Session() as sess:
    # Restore variables from disk.
    saver.restore(sess, Config.model_dir + "/model.ckpt")
    feed_dict = {X: test_x[test_example].reshape(1, 193), keep_prob:1.0}
    classification = sess.run(tf.argmax(y_,1), feed_dict)
    print("\nPrediction: " + classID[classification[0]])
```
